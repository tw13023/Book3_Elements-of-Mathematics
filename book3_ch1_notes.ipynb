### draw a heap map of 1000 digit after decimal point of pi 

```python
from mpmath import mp 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

mp.dps = 1024+1 
a_pi = mp.pi
digits = list(str(a_pi)[2:])
dig_list = list(int(x) for x in digits)
dig_array = np.array([dig_list]).reshape(32,32)
heat_map = sns.heatmap(dig_array,xticklabels=False,yticklabels=False,annot=True)
plt.show()

```

```python
from mpmath import mp
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
with st.sidebar:
    

    num_digits = st.slider('Number of decimal digits:',
                           min_value = 10000, 
                           max_value = 100000,
                           step = 10000)
    
mp.dps = num_digits + 2
pi_digits = mp.pi
# st.write(str(pi_digits))
pi_digits = str(pi_digits)[2:]
pi_digits_list = [int(x) for x in pi_digits]
# st.write(pi_digits_list)

pi_digits_array = np.array(pi_digits_list)

counts = np.bincount(pi_digits_array)

fig, ax = plt.subplots()

ax.barh(range(10), counts, align = 'center',
        edgecolor = [0.6, 0.6, 0.6])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('Count')
ax.set_ylabel('Digit, 0~9')
plt.yticks(range(10))

st.pyplot(fig)
```
